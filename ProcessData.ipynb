```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
# download file
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
#Header of downloaded file
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#open file
file1 = pd.read_csv(filename,names = headers)
```

```python
file1.head(5)
```

Steps for working with missing data:

dentify missing data
deal with missing data
correct data format

```python
#replace ? values to nan (null value)
file1.replace ("?",np.nan, inplace = True)
file1.head(10)
```

find out missing values in data file
1. isnull()
2. notnull()

```python
missing_data = file1.isnull()
print (missing_data)
```

```python
# count values with null value for each column
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Replace by mean:

"normalized-losses": 41 missing data, replace them with mean
"stroke": 4 missing data, replace them with mean
"bore": 4 missing data, replace them with mean
"horsepower": 2 missing data, replace them with mean
"peak-rpm": 2 missing data, replace them with mean
Replace by frequency:

"num-of-doors": 2 missing data, replace them with "four".
Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur
Drop the whole row:

"price": 4 missing data, simply delete the whole row
Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us



```python
# change type column normalized-losses to float
normal_mean= file1["normalized-losses"].astype(float).mean(axis = 0)
#print (normal_mean)
#apply mean to nan value
file1["normalized-losses"].replace(np.nan,normal_mean, inplace=True)
```

"stroke": 4 missing data, replace them with mean

```python
stroke_mean = round(file1["stroke"].astype(float).mean(axis=0),2)
print (stroke_mean)
file1["stroke"].replace(np.nan, stroke_mean, inplace = True)
```

```python
file1.head(10)
```

```python
# replace bore with mean
bore_mean = round(file1["bore"].astype(float).mean(axis=0),2)
file1["bore"].replace(np.nan,bore_mean, inplace = True)
```

```python
#replace them with mean "horsepower": 2 missing data, 
horsepower_mean = round(file1["horsepower"].astype(float).mean(axis = 0),2)
file1["horsepower"].replace(np.nan, horsepower_mean, inplace = True)
```

```python
#replace them with mean "peak-rpm": 2 missing data, 
peak_mean = round(file1["peak-rpm"].astype(float).mean(axis = 0),2)
file1["peak-rpm"].replace(np.nan, peak_mean,inplace = True)
```

```python
#find frequency value in num-of-doors
#file1["num-of-doors"].value_counts(): four    114 ;two      89
#"num-of-doors": 2 missing data, replace them with "four" because four takes part a larger part
file1["num-of-doors"].replace (np.nan, "four", inplace = True)
```

```python
# select row with missing data of price
price1 = file1.loc[file1["price"].isna()]
#remove whole row of missing price
file1.drop(price1.index, inplace=True)
print(price1)
```

```python
# reset index, because we droped two rows
file1.reset_index(drop=True, inplace=True)
```

Correct data format

```python
#check data type again in data file
file1.dtypes
```

```python
# update normalized-losses to int64
file1["normalized-losses"] = file1["normalized-losses"].astype("int64")
```

```python
# update Bore, Stroke, price to float type ; horsepower, and peak-rpm to int type
file1["bore"] = file1["bore"].astype("float64")
file1["stroke"] = file1["stroke"].astype("float64")
file1["horsepower"] = file1["horsepower"].astype("int64")
file1["peak-rpm"] = file1["peak-rpm"].astype("int64")
file1["price"] = file1["price"].astype("float64")
```

Data Standardization

update city-mpg; highway-mpg to standard measure
the formula for unit conversion is

L/100km = 235 / mpg

```python
file1["city-mpg"] = 235/file1["city-mpg"]
file1["highway-mpg"] = 235/file1["highway-mpg"]
file1.rename(columns = {"city-mpg":"city-L100km"}, inplace = True)
file1.rename(columns = {"highway-mpg":"highway-L100km"}, inplace = True)
```

```python
file1.head(10)
```

Data Normalisation to peak-rpm column

```python
file1["peak-rpm"].describe()
```

```python
#file1["peak-rpm"] = file1["peak-rpm"].astype("float64")
#peak_mean = file1["peak-rpm"].mean
#peak_std = file1["peak-rpm"].std
file1["peak-rpm"] = (file1["peak-rpm"]-file1["peak-rpm"].mean())/file1["peak-rpm"].std()
file1.head(10)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(file1["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# process bins for horsepower by horsepower, medium horsepower, and little horsepower (3 types)

bins = np.linspace(min(file1["horsepower"]),max(file1["horsepower"]),4)
#print (bins)
groupname = ["low", "medium", "high"]
file1["horsepower-binned"] = pd.cut(file1["horsepower"], bins, labels = groupname, include_lowest = True)
file1[["horsepower","horsepower-binned"]].head(20)
#file1["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
#plt.pyplot.hist(file1["horsepower-binned"])
pyplot.bar(groupname, file1["horsepower-binned"].value_counts())
# set x/y labels and plot title
plt.pyplot.xlabel("horsepower binned")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#make dummies for fuel type: gas, and diesel
dumdf = pd.get_dummies(file1["fuel-type"])
dumdf.rename(columns = {"fuel-type-gas":"gas","fuel-type-diesel":"diesel"}, inplace = True)
dumdf.head()
# add gas, diesel into file1
file1 = pd.concat([dumdf,file1], axis =1 )
file1.drop("fuel-type" , axis = 1, inplace = True)
```

```python
#make dummies for aspiration: turbo,std
dumdf = pd.get_dummies(file1["aspiration"])
dumdf.rename(columns = {"aspiration-turbo":"turbo","aspiration-std":"std"}, inplace = True)
dumdf.head()
# add turbo, std into file1
file1 = pd.concat([dumdf,file1], axis =1 )
file1.drop("aspiration" , axis = 1, inplace = True)
file1.head(10)
```
