```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
headers = ["sepal-length", "sepal-width", "petal-length","petal-width","class"]
#df.columns = headers
# download iris data
path = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(path, names = headers)
df.head()
```

```python
# size, shape, describe about iris data
print("Columns: ", df.columns)
print ("Shape: ", df.shape)
print ("Size: ", df.size)
print ("type of iris: ", df['class'].value_counts())
print ("--------------------")
print (df.describe())
```

```python
# divide data by 3 parts:
iris_setosa = df.loc[df['class'] == 'Iris-setosa']
iris_versicolor = df.loc[df['class'] == 'Iris-versicolor']
iris_virginica = df.loc[df['class'] == 'Iris-virginica']
```

```python
sns.FacetGrid(df, hue = "class", height =3).map(sns.distplot,"petal-length").add_legend()
sns.FacetGrid(df, hue = "class", height = 3).map(sns.distplot,"petal-width").add_legend()
sns.FacetGrid(df, hue = "class", height = 3).map(sns.distplot,"sepal-length").add_legend()
sns.FacetGrid(df, hue = "class", height = 3).map(sns.distplot,"sepal-width").add_legend()
```

```python
# for iris-setosa
counts, bin_edges = np.histogram (iris_setosa["petal-length"],bins = 10,density = True)
pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf) # blue color
plt.plot(bin_edges[1:],cdf)  # orange color
plt.show()


#for iris_virginica
counts,bin_edges=np.histogram(iris_virginica["petal-length"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()

#for iris_versicolor
counts,bin_edges=np.histogram(iris_versicolor["petal-length"],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
# Train and validate data
#data spliting
from sklearn.model_selection import train_test_split
from sklearn import datasets

array = df.values
X = array[:,0:4]
Y  = array[:,4]
validation_size = 0.2
seed = 0

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = validation_size,random_state=seed)

#print("1: " , X_train.shape, y_train.shape)
#print ("2: ", X_test.shape, y_test.shape)

num_folds = 10
num_instance = len(X_train)
scoring = 'accuracy'
```

```python
# try 5 AI algorithm to assess 
#Linear regression algorithm
#Logistic regression
#Decision tree classifier
#Gaussian Nave Base
#Support Vector Machine
```

```python
#evaluate model
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score


models = []
models.append(('LR',LogisticRegression()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('CART',DecisionTreeClassifier()))
models.append(('NB',GaussianNB()))
models.append(('SVM',SVC()))

results = []
names = []
for name, model in models:
        kfold = KFold(n_splits = num_folds, random_state = seed,shuffle = False)
        cv_results = cross_val_score(model,X_train,y_train,cv = kfold,scoring= scoring)
        results.append(cv_results)
        names.append(name)
        msg = "%s : %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print (msg)
```

```python
# evaluate score by graph
plt.figure(figsize =(15,10))
fig = plt.figure()
fig.suptitle("Evaluate algorithm scores")
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()

#LR : 0.933333 (0.081650)
#LDA : 0.966667 (0.040825)
#CART : 0.925000 (0.058333)
#NB : 0.950000 (0.055277)
#SVM : 0.958333 (0.055902)
```

```python
# Predict value by validation
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

svn = SVC()
svn.fit(X_train,y_train)
predictions = svn.predict(X_test)
print ("1-Accuracy score: ", accuracy_score (y_test,predictions))
print ("2 - Confusion matrix: ", confusion_matrix(y_test,predictions))
print ("3- Classification report: ", classification_report(y_test,predictions))
```

+ The recall means "how many of this class you find over the whole number of element of this class"

+ The precision will be "how many are correctly classified among that class"

+ The f1-score is the harmonic mean between precision & recall

+ The support is the number of occurence of the given class in your dataset (so you have 37.5K of class 0 and 37.5K of class 1, which is a really well balanced dataset.
